```python
#Altuas de hombres mexicanos Constantino Estrada Jhoselin Marisol
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt


data = loadtxt("altura5.dat")

# Cargar data a pandas DataFrame
df = pd.DataFrame(data, columns=['altura', 'grupo 1', 'grupo 2', 'grupo 3'])
df=df.round(2)

ax = df.plot.bar(x='altura', y=['grupo 1', 'grupo 2', 'grupo 3'], width=0.85)

plt.xlabel('Alturas (m)')
plt.ylabel('Frecuencia')
ax.set_title('Frecuencia de alturas hombres 1. Constantino Estrada')

print(df)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt

data = loadtxt("altura6.dat")

# Load data into a pandas DataFrame
df = pd.DataFrame(data, columns=['alturas'])
df=df.round(2)
frec=df['alturas'].value_counts().sort_index()

ax=frec.plot.bar(color='skyblue', width=0.85)
plt.xlabel('Alturas(m)')
plt.ylabel('Frecuencia')

ax.set_title('Frecuencia de alturas hombres 2. Constantino Estrada')


print(frec)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt

data = loadtxt("altura6.dat")

# Load data into a pandas DataFrame
df = pd.DataFrame(data, columns=['alturas'])
df=df.round(2)
frec=df['alturas'].value_counts()

ax=frec.plot.bar(color='orange', width=0.85)
plt.xlabel('Alturas')
plt.ylabel('Frecuencia')
ax.set_title('Frecuencia descendente de alturas hombres 2. Constantino Estrada')

print(frec)
plt.show()
```
