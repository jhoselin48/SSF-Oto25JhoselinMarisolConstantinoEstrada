```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Parámetros del sistema
omega = 1.0      # frecuencia natural
alpha = 1.0      # coeficiente no lineal
t0, tf = 0.0, 20.0

# Condiciones iniciales
x0 = 1.0        # posición inicial
v0 = 0.0        # velocidad inicial
```

```python
# Sistema de ecuaciones de primer orden
def oscilador_no_lineal(t, y):
    x, v = y
    dxdt = v
    dvdt = -omega**2 * x - alpha * x**3
    return [dxdt, dvdt]

# Vector tiempo para muestreo
t_eval = np.linspace(t0, tf, 2000)
```

```python
# Resolver la ecuación diferencial
sol = solve_ivp(
    oscilador_no_lineal,
    (t0, tf),
    [x0, v0],
    t_eval=t_eval,
    method='RK45'
)

# Extraer soluciones
t = sol.t
x = sol.y[0]
v = sol.y[1]
```

```python
# Gráfica x(t)
plt.figure(figsize=(8, 4))
plt.plot(t, x)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Oscilador armónico no lineal (término x^3)")
plt.grid()
plt.show()

# Espacio fase
plt.figure(figsize=(5, 5))
plt.plot(x, v)
plt.xlabel("x")
plt.ylabel("v")
plt.title("Espacio fase")
plt.grid()
plt.show()
```
