```python
#Altuas de hombres mexicanos Constantino Estrada Jhoselin Marisol
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt


data = loadtxt("altura5.dat")

# Cargar data a pandas DataFrame
df = pd.DataFrame(data, columns=['altura', 'grupo 1', 'grupo 2', 'grupo 3'])
df=df.round(2)

df.plot.bar(x='altura', y=['grupo 1', 'grupo 2', 'grupo 3'])

plt.xlabel('Alturas')
plt.ylabel('Frecuencia')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt

data = loadtxt("altura6.dat")

# Load data into a pandas DataFrame
df = pd.DataFrame(data, columns=['alturas'])
df=df.round(2)

frec=df['alturas'].value_counts()
frec_ordenadas=frec.sort_values(ascending=True)

frec_ordenadas.plot.bar()

plt.xlabel('Alturas')
plt.ylabel('Frecuencia')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt

data = loadtxt("altura6.dat")

# Load data into a pandas DataFrame
df = pd.DataFrame(data, columns=['alturas'])
df=df.round(2)

df['alturas'] = df['alturas'].sort_values(ascending=True)

frec=df['alturas'].value_counts()

frec.plot.bar()

plt.xlabel('Alturas')
plt.ylabel('Frecuencia')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from numpy import loadtxt

data = loadtxt("altura6.dat")

# Load data into a pandas DataFrame
df = pd.DataFrame(data, columns=['alturas'])
df=df.round(2)
frec=df['alturas'].value_counts().sort_index()

frec.plot.bar()
plt.xlabel('Alturas')
plt.ylabel('Frecuencia')

print(frec)
plt.show()
```
