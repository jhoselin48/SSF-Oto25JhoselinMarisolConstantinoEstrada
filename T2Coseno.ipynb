```python
import math
# Serie directa: cada término se calcula desde cero con potencias y factoriales
def cos_directo(x, tol):
    suma = 0.0
    k = 0
    resultados = []

    while True:
        termino = ((-1)**k) * (x**(2*k)) / math.factorial(2*k)
        suma = suma + termino

        valor_real = math.cos(x)
        if valor_real != 0:
            error_rel = abs(suma - valor_real) / abs(valor_real)
        else:
            error_rel = abs(suma - valor_real)

        resultados.append([x, k, suma, error_rel])

        if abs(termino) < tol:   # paro cuando el término ya es muy pequeño
            break

        k = k + 1

    return resultados


# Serie parcial: cada término se calcula a partir del anterior
def cos_parcial(x, tol):
    suma = 0.0
    k = 0
    termino = 1.0   # primer término
    resultados = []

    while True:
        suma = suma + termino

        valor_real = math.cos(x)
        if valor_real != 0:
            error_rel = abs(suma - valor_real) / abs(valor_real)
        else:
            error_rel = abs(suma - valor_real)

        resultados.append([x, k, suma, error_rel])

        if abs(termino) < tol:
            break

        k = k + 1
        # fórmula recursiva para obtener el siguiente término
        termino = termino * (-1) * x**2 / ((2*k-1)*(2*k))

    return resultados


# Función para imprimir la tabla
def imprimir_tabla(resultados, metodo):
    print("\nTabla método:", metodo)
    print("   x   iteración           suma        error relativo")
    for fila in resultados:
        print(f"{fila[0]:6.2f} {fila[1]:9d} {fila[2]:15.10f} {fila[3]:15.10e}")


# Programa principal
for tolerancia in [1e-4, 1e-8]:
    print("\n=== Tolerancia =", tolerancia, "===")
    for x in [0.5, 1.0, 100.0]:
        tabla1 = cos_directo(x, tolerancia)
        tabla2 = cos_parcial(x, tolerancia)

        imprimir_tabla(tabla1, "Directo, x=" + str(x))
        imprimir_tabla(tabla2, "Parcial, x=" + str(x))
```
